```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from ta import *
%matplotlib inline
```

```python
from keras.layers import Input, Dense, BatchNormalization, Dropout, LSTM, TimeDistributed, Reshape, Activation, \
    LeakyReLU, Bidirectional
from keras.optimizers import SGD, Nadam
from keras.models import Model, Sequential
from keras.metrics import categorical_accuracy

def create_sae_lastm_model(timestep, input_size, lstm_output_size):
    input_prices = Input(shape = (timestep, input_size))
    encoded1_da = Dense(hidden_layer_size, activation = 'sigmoid')(input_prices)
    encoded2_da = Dense(hidden_layer_size, activation = 'sigmoid')(encoded1_da)
    encoded3_da = Dense(hidden_layer_size, activation = 'sigmoid')(encoded2_da)
    encoded4_da = Dense(hidden_layer_size, activation = 'sigmoid')(encoded3_da)
    encoded5_da = Dense(hidden_layer_size, activation = 'sigmoid')(encoded4_da)
    decoded5_da = Dense(hidden_layer_size, activation = 'sigmoid')(encoded5_da)
    decoded_da = TimeDistributed(Dense(input_size, activation = 'sigmoid'))(decoded5_da)
    lstm = LSTM(lstm_output_size, activation = 'sigmoid')(decoded_da)
    final = Dropout(0.1)(lstm)
    dense = Dense(1, activation = 'sigmoid')(final)

    model = Model(inputs = input_prices, outputs = dense)

    model.compile(loss='mae', optimizer = 'adam', metrics=['mae'])
    
    return model
```

## Parameters

```python
input_size = 14
hidden_layer_size = 10
test_perc = 0.1
timestep = 5
lstm_output_size = 10
```

## Load data and define model

```python
df = read_data('Nifty500.csv', False)
closePriceScaler = MinMaxScaler(feature_range=(0, 1))
X_train, X_test, Y_train, Y_test = split_dataset(df, closePriceScaler, test_perc, timestep)
model = create_sae_lastm_model(timestep, input_size, lstm_output_size)
```

```python
plt.plot(np.append(Y_train, Y_test), 'b')
plt.ylabel('Price ($)')
plt.show()
```

## Train

```python
model.fit(X_train, Y_train,
                 epochs = 1500, batch_size = 1000,
                 validation_data=(X_test, Y_test),
                 shuffle = True)
```

## Visualize result

```python
pred = model.predict(X_test)

MAE, MeAE, R2 = genereate_metrics(Y_test, pred)

# pred = transform_to_original(pred, closePriceScaler)
# y = transform_to_original(Y_test, closePriceScaler)

print('MAE: ', MAE)
print('MeAE: ', MeAE)
print('R2: ', R2)

plt.plot(pred, 'b')
plt.plot(Y_test, 'r')
plt.ylabel('Price ($)')
plt.show()
```

## Preprocessing

```python
def read_data(fileName, floats):
    data = pd.read_csv(fileName)
    if not floats:
        data_proc = [[float(x.replace(",", "")) for x in reversed(data['Price'].values)],
                     [float(x.replace(",", "")) for x in reversed(data['High'].values)],
                     [float(x.replace(",", "")) for x in reversed(data['Low'].values)]]
    else:
        data_proc = [[x for x in reversed(data['Price'].values)],
                     [x for x in reversed(data['High'].values)],
                     [x for x in reversed(data['Low'].values)]]

    df = pd.DataFrame(data=np.array(data_proc).T.tolist(), columns=['Close', 'High', 'Low'])
    df = utils.dropna(df)

    df['MACD'] = macd(df['Close'], n_fast=12, n_slow=26, fillna=True)
    df['CCI'] = cci(df['High'], df['Low'], df['Close'], n=20, c=0.015, fillna=True)
    df['ATR'] = average_true_range(df['High'], df['Low'], df['Close'], n=14, fillna=True)
    df['BOLL'] = bollinger_hband(df["Close"], n=20, ndev=2, fillna=True)
    df['EMA20'] = ema(df["Close"], n=20, fillna=True)
    df['SMA'] = stoch_signal(df['High'], df['Low'], df['Close'], n=14, d_n=5, fillna=True)
    df['AwesomeOscilator'] = ao(df['High'], df['Low'], s=5, l=34, fillna=True)
    df['RSI'] = rsi(df["Close"], n=14, fillna=True)
    df['StohOscilator'] = stoch(df['High'], df['Low'], df['Close'], n=14, fillna=True)
    df['UltimateOscilator'] = uo(df['High'], df['Low'], df['Close'], s=7, m=14, l=28, ws=4.0, wm=2.0, wl=1.0,
                                 fillna=True)
    df['WR'] = wr(df['High'], df['Low'], df['Close'], lbp=14, fillna=True)

    return df
```

```python
def plot(var):
    plt.plot(var, 'b')
    plt.ylabel('Price ($)')
    plt.show()
```

```python
def split_dataset(df, closePriceScaler, test_perc=0.3, timestep=5):
    X = np.array(df.values)
    nr_train = int(len(X.T[0]) * (1 - test_perc))
    Y = X.T[0]

    scaler = MinMaxScaler(feature_range=(0, 1))
    X = scaler.fit_transform(X)
    
    Y = closePriceScaler.fit_transform(np.array(Y).reshape(-1, 1))

    Y_train = Y[1 + timestep:nr_train + 1]
    Y_test = Y[nr_train + 1:]

    X = X[:-1]

    X = [[X[batch_nr - t] for t in reversed(range(timestep))] for batch_nr in range(timestep, len(X))]

    X = np.array(X)
    nr_train -= timestep
    X_train = X[0:nr_train]
    X_test = X[nr_train:]

    return X_train, X_test, Y_train, Y_test
```

```python
def split_dataset_random(df, closePriceScaler, test_perc=0.3, timestep=5):
    X = np.array(df.values)
    nr_train = int(len(X.T[0]) * (1-test_perc))  
    Y = X.T[0]


    scaler = MinMaxScaler(feature_range=(0, 1))
    X = scaler.fit_transform(X)

    X = X[:-1]
    X = [[X[batch_nr - t] for t in reversed(range(timestep))] for batch_nr in range(timestep, len(X))]
    X = np.array(X)

    Y = closePriceScaler.fit_transform(np.array(Y).reshape(-1,1))
    Y = Y[timestep+1:]

    p = np.random.permutation(X.shape[0])
    X = X[p]
    Y=Y[p]

    nr_train = int(X.shape[0] * (1-test_perc))  

    Y_train = Y[0:nr_train]
    Y_test = Y[nr_train:]

    X_train = X[0:nr_train]
    X_test = X[nr_train:]

    return X_train, X_test, Y_train, Y_test
```

```python
from sklearn import metrics

def genereate_metrics(y, pred):
    MAE = metrics.mean_absolute_error(y, pred)
    MeAE = metrics.median_absolute_error(y, pred)
    R2 = metrics.r2_score(y, pred)  
    return MAE,MeAE, R2
```

```python
def transform_to_original(val, closePriceScaler):
    return closePriceScaler.inverse_transform(val.reshape(-1,1)).ravel() 
```

```python
def test_profitability(y, pred, B = -0.0001, S=-0.0001):
    R = 0

    good_deals = 0
    bad_deals = 0
    bah = ((y[-1] - y[0]) / y[0]) * 100

    for i in range(len(y)-1):
        add = 0
        if pred[i] > y[i]:
            add += (y[i+1] - y[i] + y[i] * B + y[i+1]*S)/y[i]
        else:
            add += (y[i] - y[i+1] + y[i+1] * B + y[i]*S)/y[i]

        if add > 0:
            good_deals += 1 
        else:
            bad_deals +=1
        R += add

    R *= 100
    
    return R, good_deals, bad_deals, bah
```
